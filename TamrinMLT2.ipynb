```python
img0 = cv2.imread(f'{data_path}/0.png',0)
img1 = cv2.imread(f'{data_path}/1.png',0)
img2 = cv2.imread(f'{data_path}/2.png',0)
img3 = cv2.imread(f'{data_path}/3.png',0)
img4 = cv2.imread(f'{data_path}/4.png',0)
img5 = cv2.imread(f'{data_path}/5.png',0)
img6 = cv2.imread(f'{data_path}/6.png',0)
img7 = cv2.imread(f'{data_path}/7.png',0)
img8 = cv2.imread(f'{data_path}/8.png',0)
img9 = cv2.imread(f'{data_path}/9.png',0)


_, binary_image = cv2.threshold(img0, 127, 255, cv2.THRESH_BINARY)
# Invert the binary image
img0 = cv2.bitwise_not(binary_image)

kernel_size = 2
kernel = np.ones((kernel_size, kernel_size), np.uint8)
img0 = cv2.dilate(img0, kernel, iterations=1)



_, binary_image = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY)
# Invert the binary image
img1 = cv2.bitwise_not(binary_image)

kernel_size = 2
kernel = np.ones((kernel_size, kernel_size), np.uint8)
img1 = cv2.dilate(img1, kernel, iterations=1)



_, binary_image = cv2.threshold(img2, 127, 255, cv2.THRESH_BINARY)
# Invert the binary image
img2 = cv2.bitwise_not(binary_image)



_, binary_image = cv2.threshold(img3, 155, 255, cv2.THRESH_BINARY)
# Invert the binary image
img3 = cv2.bitwise_not(binary_image)



_, binary_image = cv2.threshold(img4, 180, 255, cv2.THRESH_BINARY)
# Invert the binary image
img4 = cv2.bitwise_not(binary_image)

_, binary_image = cv2.threshold(img5, 150, 255, cv2.THRESH_BINARY)
# Invert the binary image
img5 = cv2.bitwise_not(binary_image)

_, binary_image = cv2.threshold(img6, 190, 255, cv2.THRESH_BINARY)
# Invert the binary image
img6 = cv2.bitwise_not(binary_image)



_, binary_image = cv2.threshold(img7, 170, 255, cv2.THRESH_BINARY)
# Invert the binary image
img7 = cv2.bitwise_not(binary_image)
kernel_size = 2
kernel = np.ones((kernel_size, kernel_size), np.uint8)
img7 = cv2.dilate(img7, kernel, iterations=1)



_, binary_image = cv2.threshold(img8, 150, 255, cv2.THRESH_BINARY)
# Invert the binary image
img8 = cv2.bitwise_not(binary_image)
kernel_size = 2
kernel = np.ones((kernel_size, kernel_size), np.uint8)
img8 = cv2.dilate(img8, kernel, iterations=1)




_, binary_image = cv2.threshold(img9, 200, 255, cv2.THRESH_BINARY)
# Invert the binary image
img9 = cv2.bitwise_not(binary_image)



# Show the Images
plt.figure(figsize=[20, 10])  # Adjust the figure size as needed

plt.subplot(2, 5, 1); plt.imshow(img0, cmap='gray'); plt.title("Fig0")
plt.subplot(2, 5, 2); plt.imshow(img1, cmap='gray'); plt.title("Fig1")
plt.subplot(2, 5, 3); plt.imshow(img2, cmap='gray'); plt.title("Fig2")
plt.subplot(2, 5, 4); plt.imshow(img3, cmap='gray'); plt.title("Fig3")
plt.subplot(2, 5, 5); plt.imshow(img4, cmap='gray'); plt.title("Fig4")
plt.subplot(2, 5, 6); plt.imshow(img5, cmap='gray'); plt.title("Fig5")
plt.subplot(2, 5, 7); plt.imshow(img6, cmap='gray'); plt.title("Fig6")
plt.subplot(2, 5, 8); plt.imshow(img7, cmap='gray'); plt.title("Fig7")
plt.subplot(2, 5, 9); plt.imshow(img8, cmap='gray'); plt.title("Fig8")
plt.subplot(2, 5, 10); plt.imshow(img9, cmap='gray'); plt.title("Fig9")

plt.tight_layout()  # Adjusts subplots to fit into the figure area.
plt.show()


```
